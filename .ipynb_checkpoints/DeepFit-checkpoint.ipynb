```python
device = 'cpu'

model = NeuralNetwork().to(device)
```

```python
learning_rate = 1e-3
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
loss_fn = torch.nn.MSELoss()
```

```python
for t in range(1000):
    n_pix = 1000
    X = torch.rand(1, 1, n_pix, n_pix, device=device)
    X[0,0,20,20]+= 5
    X[0,0,21,20]+= 3
    X[0,0,20,21]+= 3
    X[0,0,20,19]+= 3
    X[0,0,19,20]+= 3
    X[0,0,19,21]+= 1
    X[0,0,21,21]+= 1
    X[0,0,21,19]+= 1
    X[0,0,19,19]+= 1
    alpha_est = model(X)
    m0, alpha, w_map = legacyMTT.wilks_map(X.view(n_pix, -1).detach().numpy(), 1, 7) 
    loss = loss_fn(alpha_est, torch.from_numpy(alpha).view(1,1,n_pix,-1).type(torch.float))
    if t % 100 == 1:
        print(t, loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.imshow(alpha, cmap='gray')
```

```python
plt.imshow(alpha_est.detach().view(n_pix,-1), cmap='gray')
```

```python
sigma = torch.tensor([80.]) #nm The positional uncertainty of a photon due to diffraction. 0.42 lambda / 2 / NA with lambda about 600nm and NA 1.4
sigma_px = sigma/160. #Assume pixel size of 160nm
mean_num_photons = 64 #Depends on illumination intensity and camera integration time. This value gives a similar SNR to typical experimental data.

r = 1000
c = 1000

x_c = np.random.rand()*r
y_c = np.random.rand()*c

print(x_c)
print(y_c)
```

```python
num_emitted_photons = dist.poisson.Poisson(mean_num_photons).sample()

print(num_emitted_photons)
```

```python
position_dist = dist.multivariate_normal.MultivariateNormal(torch.tensor([x_c, y_c]), covariance_matrix=torch.eye(2)*sigma**2)
```

```python
positions = position_dist.sample((num_emitted_photons.int(),))
```
